## Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

````python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
````

### PorterStemmer

````python
from nltk.stem import PorterStemmer
````

````python
stemming=PorterStemmer()
````

````python
for word in words:
    print(word+"---->"+stemming.stem(word))
````

````python
stemming.stem('congratulations')
````

````python
stemming.stem("sitting")
````

## Stemming using `PorterStemmer`

**Stemming** is the process of reducing words to their **root (stem)** by removing prefixes or suffixes.

It helps treat different forms of a word as the same word during text processing.

### Example

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

words = [
    "eating", "eats", "eaten",
    "writing", "writes",
    "programming", "programs",
    "history", "finally", "finalized"
]

for word in words:
    print(word + " ---> " + stemming.stem(word))
```

### Concepts Used

- `PorterStemmer()` → Creates a stemming object.
- `stem(word)` → Returns the stem (root form) of a word.
- `for` loop → Applies stemming to every word in the list.

### Example Output

```text
eating       ---> eat
eats         ---> eat
eaten        ---> eaten
writing      ---> write
writes       ---> write
programming  ---> program
programs     ---> program
history      ---> histori
finally      ---> final
finalized    ---> final
```

> **Remember:** Stemming removes prefixes/suffixes using rules, so the resulting word **may not be a valid English word** (e.g., `history → histori`).

---

## Why is Stemming Used?

Suppose you want to classify product reviews as **Positive** or **Negative**.

Reviews:

```text
I am eating pizza.
He eats pizza daily.
She has eaten pizza.
```

Without stemming, the model sees:

```text
eating
eats
eaten
```

as **three different words**.

After stemming:

```text
eat
eat
eat (or similar root)
```

they are treated as the **same concept**, reducing vocabulary size and improving model performance.

---

## Note

- **Stemming** → Fast, rule-based, may produce non-dictionary words.
- Used in **search engines**, **text classification**, and **information retrieval**.
- Later, you'll learn **Lemmatization**, which produces proper dictionary words and is generally more accurate.

### RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

````python
from nltk.stem import RegexpStemmer
````

````python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
````

````python
reg_stemmer.stem('eating')
````

````python
reg_stemmer.stem('ingeating')
````

````python
reg_stemmer.stem('manageable')
````

## `RegexpStemmer` using NLTK

`RegexpStemmer` performs **stemming using regular expressions (regex)**. Instead of predefined stemming rules (like PorterStemmer), you define the patterns to remove.

### Example

```python
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)

reg_stemmer.stem('eating')
reg_stemmer.stem('ingeating')
```

### Concepts Used

- `RegexpStemmer()` → Creates a regex-based stemmer.
- `'ing$|s$|e$|able$'` → Regex pattern specifying suffixes to remove.
- `$` → Matches the **end of the word**.
- `|` → Means **OR**.
- `min=4` → Applies stemming only if the word length is **4 or more**.
- `stem(word)` → Returns the stemmed word.

### Example Output

```python
reg_stemmer.stem("eating")
# Output: 'eat'

reg_stemmer.stem("ingeating")
# Output: 'ingeat'
```

> **Remember:** `RegexpStemmer` removes **only the suffixes you specify** using regular expressions. It does not automatically learn or apply linguistic rules like `PorterStemmer`.

## `SnowballStemmer` using NLTK

`SnowballStemmer` is an improved version of `PorterStemmer`. It provides **more accurate and consistent stemming** and supports **multiple languages**.

### Example

```python
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer("english")

for word in words:
    print(word + " ---> " + snowball_stemmer.stem(word))
```

### Concepts Used

- `SnowballStemmer("english")` → Creates a stemmer for the English language.
- `stem(word)` → Returns the stem (root form) of a word.
- Supports multiple languages such as English, French, German, Spanish, etc.

### Example Output

```text
eating       ---> eat
eats         ---> eat
eaten        ---> eaten
writing      ---> write
writes       ---> write
programming  ---> program
programs     ---> program
history      ---> histori
finally      ---> final
finalized    ---> final
```

### Comparison

```python
stemming.stem("fairly")
# Output: fairli

snowball_stemmer.stem("fairly")
# Output: fair
```

```python
stemming.stem("sportingly")
# Output: sportingli

snowball_stemmer.stem("sportingly")
# Output: sport
```

```python
stemming.stem("goes")
# Output: goe

snowball_stemmer.stem("goes")
# Output: goe
```

> **Remember:** `SnowballStemmer` generally produces **better stems** than `PorterStemmer`, but it is still a **rule-based stemmer**, so the output may not always be a valid dictionary word.

## Difference Between Stemmers

| Stemmer | Rules | Accuracy | Languages | Best Use |
|---------|-------|----------|-----------|----------|
| **PorterStemmer** | Fixed rules | Good | English | Basic stemming |
| **RegexpStemmer** | User-defined regex | Depends on regex | Any | Custom suffix removal |
| **SnowballStemmer** | Improved rules | Better than Porter | Multiple | Recommended rule-based stemmer |

> **Interview Tip:** If asked to choose a rule-based stemmer, **SnowballStemmer** is generally preferred over **PorterStemmer** because it produces more linguistically meaningful stems.
