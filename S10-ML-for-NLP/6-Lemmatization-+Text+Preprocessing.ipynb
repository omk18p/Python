## Wordnet Lemmatizer
Lemmatization technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example −


````python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')      # Required for newer NLTK versions
nltk.download('wordnet')
nltk.download('omw-1.4')        # WordNet multilingual data
nltk.download('stopwords')
````

````python
## Q&A,chatbots,text summarization
from nltk.stem import WordNetLemmatizer
````

````python
lemmatizer=WordNetLemmatizer()
````

````python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='v')
````

````python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
````

## Lemmatization using `WordNetLemmatizer`

**Lemmatization** is the process of converting a word to its **base (dictionary) form**, called a **lemma**.

Unlike stemming, lemmatization uses a **dictionary (WordNet)** and the **Part of Speech (POS)** to produce meaningful words.

### Example

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatizer.lemmatize("going", pos='v')
```

### POS Tags

| POS | Meaning |
|-----|---------|
| `'n'` | Noun |
| `'v'` | Verb |
| `'a'` | Adjective |
| `'r'` | Adverb |

### Concepts Used

- `WordNetLemmatizer()` → Creates a lemmatizer object.
- `lemmatize(word, pos)` → Returns the dictionary (base) form of the word.
- `pos='v'` → Treats the word as a **verb**, leading to more accurate results.

### Example Output

```python
lemmatizer.lemmatize("going", pos='v')
# Output: 'go'

lemmatizer.lemmatize("goes", pos='v')
# Output: 'go'
```

```python
for word in words:
    print(word + " ---> " + lemmatizer.lemmatize(word, pos='v'))
```

Output


### Example

```python
lemmatizer.lemmatize("fairly", pos='v')
# Output: fairly

lemmatizer.lemmatize("sportingly")
# Output: sportingly
```

> **Remember:** The correct **POS tag** is important. Without it, the lemmatizer may not return the expected base form.

## Stemming vs Lemmatization

| Stemming | Lemmatization |
|----------|---------------|
| Rule-based | Dictionary-based (WordNet) |
| Faster | Slightly slower |
| May produce invalid words | Produces valid dictionary words |
| Doesn't use POS | Uses POS tags |
| Example: `history → histori` | `history → history` |
| Example: `finalized → final` | `finalized → finalize` |

> **Interview Tip ⭐:** If asked which is better, **Lemmatization** is generally preferred because it returns **meaningful dictionary words**, while **Stemming** is faster but may produce incorrect or incomplete words.
