## Stopwords

**Stopwords** are common words that appear frequently in a language but usually add little meaning to text analysis.

### Examples

```text
the, is, are, am, was, were, and, of, in, to
```

### Why Remove Stopwords?

- Reduces noise
- Decreases vocabulary size
- Improves NLP model efficiency

### NLTK

```python
from nltk.corpus import stopwords

stopwords.words("english")
```

> **Remember:** Stopwords are removed in many NLP tasks, but they may be kept when sentence meaning is important (e.g., machine translation, question answering, LLMs).

## Stopword Removal + Stemming

This code preprocesses each sentence by:
1. Tokenizing the sentence into words.
2. Removing English stopwords.
3. Applying stemming to the remaining words.
4. Joining the processed words back into a sentence.

### Code Flow

```text
Sentence
    │
    ▼
Word Tokenization
    │
    ▼
Remove Stopwords
    │
    ▼
Apply Stemming
    │
    ▼
Join Words → Clean Sentence
```

### Concepts Used

- `range(len(sentences))` → Iterates through each sentence by index.
- `word_tokenize()` → Splits a sentence into words (tokens).
- `stopwords.words('english')` → Returns English stopwords.
- `set()` → Makes stopword lookup faster (`O(1)` average time).
- `stemmer.stem(word)` → Converts words to their stem.
- `list comprehension` → Performs filtering and stemming in one line.
- `' '.join(words)` → Converts the list of processed words back into a sentence.

### Example

Original Sentence:

```text
I am learning Natural Language Processing.
```

After Stopword Removal:

```text
learning Natural Language Processing
```

After Stemming:

```text
learn natur languag process
```

> **Remember:** This preprocessing reduces noise and vocabulary size, making the text suitable for machine learning models.

## Stopword Removal + Snowball Stemming

This code preprocesses each sentence by:
1. Tokenizing the sentence into words.
2. Removing English stopwords.
3. Applying **Snowball Stemming** to the remaining words.
4. Joining the processed words back into a sentence.

### Code Flow

```text
Sentence
    │
    ▼
Word Tokenization
    │
    ▼
Remove Stopwords
    │
    ▼
Apply Snowball Stemming
    │
    ▼
Join Words → Clean Sentence
```

### Concepts Used

- `word_tokenize()` → Splits a sentence into words.
- `stopwords.words('english')` → Returns English stopwords.
- `set()` → Faster stopword lookup.
- `snowballstemmer.stem(word)` → Converts words to improved stems.
- `list comprehension` → Filters and stems words in one step.
- `' '.join(words)` → Converts the processed words back into a sentence.

### Example

Original Sentence:

```text
I am learning Natural Language Processing.
```

After Stopword Removal:

```text
learning Natural Language Processing
```

After Snowball Stemming:

```text
learn natur languag process
```

> **Remember:** SnowballStemmer is an improved version of PorterStemmer. It generally produces more accurate stems and is the preferred rule-based stemmer.

````python
sentences=nltk.sent_tokenize(paragraph)
````

````python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
````

````python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
````

## Stopword Removal + Lemmatization

This code preprocesses each sentence by:
1. Tokenizing the sentence into words.
2. Removing English stopwords.
3. Converting words to lowercase.
4. Applying **Lemmatization**.
5. Joining the processed words back into a sentence.

### Code Flow

```text
Sentence
    │
    ▼
Word Tokenization
    │
    ▼
Remove Stopwords
    │
    ▼
Convert to Lowercase
    │
    ▼
Apply Lemmatization
    │
    ▼
Join Words → Clean Sentence
```

### Concepts Used

- `word_tokenize()` → Splits a sentence into words.
- `stopwords.words('english')` → Returns English stopwords.
- `set()` → Faster stopword lookup.
- `word.lower()` → Converts each word to lowercase.
- `lemmatizer.lemmatize(word, pos='v')` → Converts a word to its dictionary (base) form, treating it as a **verb**.
- `list comprehension` → Removes stopwords and lemmatizes in one step.
- `' '.join(words)` → Converts the processed words back into a sentence.

### Example

Original Sentence:

```text
I am Learning Natural Language Processing.
```

After Stopword Removal:

```text
Learning Natural Language Processing
```

After Lowercase + Lemmatization:

```text
learn natural language process
```

> **Remember:** Unlike stemming, **lemmatization returns meaningful dictionary words**, making it more accurate for most NLP applications.

````python
sentences=nltk.sent_tokenize(paragraph)
````

````python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
````

````python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
````
