## Sentence Tokenization using NLTK

Sentence tokenization splits a paragraph into **individual sentences**.

### Example

```python
from nltk.tokenize import sent_tokenize

corpus = """
Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""

documents = sent_tokenize(corpus)

print(documents)
```

### Concepts Used

- `sent_tokenize()` → Splits text into sentences.
- `documents` → A list containing all the sentences.
- `type(documents)` → Returns `<class 'list'>`.
- `for sentence in documents:` → Iterates through each sentence.
- `print(sentence)` → Prints one sentence at a time.

### Output

```python
[
"Hello Welcome,to Krish Naik's NLP Tutorials.",
"Please do watch the entire course!",
"to become expert in NLP."
]
```

> **Remember:** `sent_tokenize()` returns a **list of sentences**, not words.

````python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
````

````python
word_tokenize(corpus)
````

````python
for sentence in documents:
    print(word_tokenize(sentence))
````

## Word Tokenization using NLTK

Word tokenization splits a **paragraph or sentence into individual words (tokens)**.

### Example

```python
from nltk.tokenize import word_tokenize

word_tokenize(corpus)

for sentence in documents:
    print(word_tokenize(sentence))
```

### Concepts Used

- `word_tokenize()` → Splits text into words and punctuation (tokens).
- `word_tokenize(corpus)` → Tokenizes the entire paragraph.
- `for sentence in documents:` → Loops through each sentence.
- `word_tokenize(sentence)` → Tokenizes each sentence separately.

### Output

```python
['Hello', 'Welcome', ',', 'to', 'Krish', 'Naik', "'s", 'NLP', 'Tutorials', '.']

['Please', 'do', 'watch', 'the', 'entire', 'course', '!']

['to', 'become', 'expert', 'in', 'NLP', '.']
```

> **Remember:** `word_tokenize()` returns a **list of tokens**, where punctuation marks are also treated as separate tokens.

````python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
````

## `wordpunct_tokenize()` using NLTK

`wordpunct_tokenize()` splits text into **words and punctuation**, treating punctuation as separate tokens.

### Example

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

### Concepts Used

- `wordpunct_tokenize()` → Splits text based on **words and punctuation**.
- Punctuation marks are returned as separate tokens.
- Apostrophes (`'`) are also split into separate tokens.

### Output

```python
['Hello', 'Welcome', ',', 'to', 'Krish', 'Naik', "'", 's', 'NLP', 'Tutorials', '.', 'Please', 'do', 'watch', 'the', 'entire', 'course', '!', 'to', 'become', 'expert', 'in', 'NLP', '.']
```

> **Remember:** Unlike `word_tokenize()`, `wordpunct_tokenize()` splits words containing apostrophes or punctuation into separate tokens (e.g., `"Naik's"` → `['Naik', "'", 's']`).

## `TreebankWordTokenizer` using NLTK

`TreebankWordTokenizer` tokenizes text using the **Penn Treebank tokenization rules**. It is commonly used in NLP because it handles contractions and punctuation more naturally.

### Example

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```

### Concepts Used

- `TreebankWordTokenizer()` → Creates a tokenizer object.
- `tokenizer.tokenize(corpus)` → Splits the text into tokens using Treebank rules.
- Keeps contractions together where appropriate (e.g., `"Naik's"` → `["Naik", "'s"]`).

### Output

```python
['Hello', 'Welcome', ',', 'to', 'Krish', 'Naik', "'s", 'NLP', 'Tutorials.', 'Please', 'do', 'watch', 'the', 'entire', 'course', '!', 'to', 'become', 'expert', 'in', 'NLP', '.']
```

> **Remember:** `TreebankWordTokenizer` is more linguistically accurate than `wordpunct_tokenize()`. It keeps contractions like `"'s"` together instead of splitting them into `"'"` and `"s"`.

## Difference Between NLTK Tokenizers

| Tokenizer | Splits Into | Handles Contractions | Punctuation | Best Use |
|-----------|-------------|----------------------|-------------|----------|
| `sent_tokenize()` | Sentences | N/A | Keeps punctuation within sentences | Split a paragraph into sentences |
| `word_tokenize()` | Words + punctuation | ✅ Keeps contractions together (`"'s"`) | Separate tokens | General-purpose word tokenization |
| `wordpunct_tokenize()` | Words and punctuation | ❌ Splits contractions (`"'"`, `"s"`) | Separate tokens | Simple punctuation-based tokenization |
| `TreebankWordTokenizer()` | Words + punctuation | ✅ Keeps contractions together | Uses Penn Treebank rules | More linguistically accurate tokenization |

### Example

Text:

```text
Krish Naik's NLP Tutorials.
```

| Tokenizer | Output |
|-----------|--------|
| `word_tokenize()` | `['Krish', 'Naik', "'s", 'NLP', 'Tutorials', '.']` |
| `wordpunct_tokenize()` | `['Krish', 'Naik', "'", 's', 'NLP', 'Tutorials', '.']` |
| `TreebankWordTokenizer()` | `['Krish', 'Naik', "'s", 'NLP', 'Tutorials', '.']` |

> **Remember:**
> - `sent_tokenize()` → Paragraph ➜ Sentences
> - `word_tokenize()` → Standard word tokenizer (most commonly used)
> - `wordpunct_tokenize()` → Splits every punctuation separately
> - `TreebankWordTokenizer()` → More accurate tokenizer following Penn Treebank rules
