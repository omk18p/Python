## Variable-Length Arguments (`*args`)

Used when a function can accept **any number of positional arguments**.

### Syntax

```python
def function_name(*args):
```

- `*` → Collects multiple positional arguments.
- `args` → Tuple containing all arguments.
- `args` is just a variable name (can be anything).

### Example

```python
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(10, 20, 30)
```

Output

```python
10
20
30
```

### Remember

- `*args` → Variable-length **positional** arguments (**tuple**).
- `**kwargs` → Variable-length **keyword** arguments (**dictionary**).

````python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*args):
    for number in args:
        print(number)
````

````python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
````

````python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
````

````python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
````

## Keyword Arguments (`**kwargs`)

Used when a function can accept **any number of keyword arguments**.

### Syntax

```python
def function_name(**kwargs):
```

- `**` → Collects multiple keyword arguments.
- `kwargs` → Dictionary containing all key-value pairs.
- `kwargs` is just a variable name (can be anything).

### Example

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Krish", age=32, country="India")
```

Output

```python
name: Krish
age: 32
country: India
```

### Remember

- `*args` → Variable-length **positional** arguments (**tuple**).
- `**kwargs` → Variable-length **keyword** arguments (**dictionary**).

````python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
````

````python
print_details(name="Krish",age="32",country="India")
````

````python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
````

````python
print_details(1,2,3,4,"Krish",name="Krish",age="32",country="India")
````

## Returning Multiple Values

A function can return **multiple values**.

```python
return value1, value2
```

Python returns them as a **tuple**.

```python
def multiply(a, b):
    return a * b, a
```

```python
product, num = multiply(2, 3)
```

- `product = 6`
- `num = 2`
