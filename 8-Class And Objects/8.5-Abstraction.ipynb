#### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

````python
from abc import ABC,abstractmethod

## Abstract base cclass
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)

````

## Abstraction with Polymorphism

**Abstraction** defines a common interface, while **polymorphism** allows different child classes to implement that interface in their own way.

### Key Points

- `Vehicle` → Abstract Base Class (`ABC`); cannot be instantiated.
- `@abstractmethod start_engine()` → Declares a method that **must** be implemented by every child class.
- `drive()` → Concrete (implemented) method that is inherited by all child classes.
- `Car` → Inherits from `Vehicle` and provides its own implementation of `start_engine()`.
- `operate_vehicle(vehicle)` → Accepts any object derived from `Vehicle`.
- `vehicle.start_engine()` → Executes the child class implementation (**runtime polymorphism**).
- `vehicle.drive()` → Executes the common method inherited from `Vehicle`.
- New vehicle types (e.g., `Bike`, `Truck`) can be added by implementing `start_engine()` without modifying `operate_vehicle()`.

### Advantages

- Enforces a common interface for all child classes.
- Promotes code reuse through inherited methods.
- Improves flexibility, scalability, and maintainability.
- Reduces code duplication and makes adding new classes easier.

### Output

```python
Car enginer started
The vehicle is used for driving
```

> **Remember:**  
> **Abstraction** → Defines **what** child classes must implement.  
> **Polymorphism** → Defines **how** each child class implements it.
