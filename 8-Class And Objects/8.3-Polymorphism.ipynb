#### Polymorphism
Polymorphism is a core concept in Object-Oriented Programming (OOP) that allows objects of different classes to be treated as objects of a common superclass. It provides a way to perform a single action in different forms. Polymorphism is typically achieved through method overriding and interfaces

###  Method Overriding
Method overriding allows a child class to provide a specific implementation of a method that is already defined in its parent class.

````python
## Base Class
class Animal:
    def speak(self):
        return "Sound of the animal"
    
## Derived Class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"
    
## Derived class
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
## Function that demonstrates polymorphism
def animal_speak(animal):
    print(animal.speak())
    
dog=Dog()
cat=Cat()
print(dog.speak())
print(cat.speak())
animal_speak(dog)
animal_speak(cat)

````

## Polymorphism

**Polymorphism** means **"many forms."** It allows the **same method name** to have **different behaviors** depending on the object that calls it.

### Key Points

- Achieved using **method overriding**.
- A **parent class** defines a common method.
- **Child classes** provide their own implementation of that method.
- The same function can work with different object types.
- Python decides which method to call **at runtime** (Runtime Polymorphism).

### Code Explanation

- `Animal` → Parent class with the common `speak()` method.
- `Dog` and `Cat` → Child classes that override `speak()`.
- `animal_speak(animal)` → Accepts any object that has a `speak()` method.
- When a `Dog` object is passed, `Dog.speak()` is called.
- When a `Cat` object is passed, `Cat.speak()` is called.

### Advantages

- Improves **code reusability**.
- Makes code **flexible** and **easy to extend**.
- Reduces the need for multiple `if-else` conditions.

### Output

```python
Woof!
Meow!
Woof!
Meow!
```

> **Remember:**  
> **Polymorphism = One interface (same method), many implementations (different behavior).**

````python
### Polymorphissm with Functions and MEthods
## base class
class Shape:
    def area(self):
        return "The area of the figure"
    
## Derived class 1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width * self.height
    
##DErived class 2

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius *self.radius
    
## Fucntion that demonstrates polymorphism

def print_area(shape):
    print(f"the area is {shape.area()}")


rectangle=Rectangle(4,5)
circle=Circle(3)

print_area(rectangle)
print_area(circle)



````

## Polymorphism with Functions and Methods

Polymorphism allows the **same function or method** to work with **different object types**, where each object provides its own implementation.

### Key Points

- `Shape` → Parent class with the common `area()` method.
- `Rectangle` and `Circle` → Child classes that override `area()`.
- `print_area(shape)` → Accepts any object that has an `area()` method.
- `shape.area()` → Calls the correct method based on the object's type (**runtime polymorphism**).

### Advantages

- Improves **code reusability**.
- Makes code **flexible** and **easy to extend**.
- Eliminates the need for multiple `if-else` statements.

### Output

```python
The area is 20
The area is 28.26
```

> **Remember:** The same function can work with different objects because each object implements the same method in its own way.

#### Polymorphism with Abstract Base Classes
Abstract Base Classes (ABCs) are used to define common methods for a group of related objects. They can enforce that derived classes implement particular methods, promoting consistency across different implementations.

````python
from abc import ABC,abstractmethod

## Define an abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

## Derived class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car enginer started"
    
## Derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle enginer started"
    
# Function that demonstrates polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())

## create objects of cAr and Motorcycle

car = Car()
motorcycle = Motorcycle()

start_vehicle(car)


````

It enables a single function to handle objects of different classes, each with its own implementation of a method. By understanding and applying polymorphism, you can create more extensible and maintainable object-oriented programs.

## Abstraction

**Abstraction** hides implementation details and defines **what a class must do**, not **how** it does it.

### Key Points

- `ABC` → Used to create an **Abstract Base Class**.
- `@abstractmethod` → Declares a method that **must be implemented** by every child class.
- An **abstract class cannot be instantiated** (no objects can be created).
- Child classes **must override** all abstract methods, otherwise their objects cannot be created.
- `pass` → Placeholder; provides no implementation.
- `Car` and `Motorcycle` → Implement `start_engine()` in their own way.
- `start_vehicle(vehicle)` → Accepts any object derived from `Vehicle`.
- Python calls the correct `start_engine()` method based on the object's type (**runtime polymorphism**).

### Advantages

- Enforces a common interface.
- Improves code consistency.
- Makes code easier to extend and maintain.

### Output

```python
Car enginer started
```

> **Remember:**  
> **Abstraction** = Defines **what** must be implemented.  
> **Polymorphism** = Defines **how** different objects implement the same method.
