##### Importing Modules in Python: Modules and Packages
In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

````python
import math
math.sqrt(17)
````

````python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(pi))
print(pi)
````

````python
import numpy as np
np.array([1,2,3,4])
````

````python
from math import *
print(sqrt(16))
print(pi)
````

````python
from package.maths import addition
addition(2,3)
````

````python
from package import maths
maths.addition(2,3)
````

## Importing from a Package

Used to import a **specific function** from a module inside a package.

### Syntax

```python
from package.module import function
```

### Example

```python
from package.maths import addition

addition(2, 3)
```

- `package` → Folder containing Python modules.
- `maths` → Module (`maths.py`).
- `addition` → Function imported from the module.
- After importing, call the function directly.

### `__init__.py`

`__init__.py` is a special file that **marks a folder as a Python package**.

Example:

```text
package/
│── __init__.py
│── maths.py
│── main.py
```

- Helps Python recognize the folder as a package.
- Can contain package initialization code (optional).
- Can also export functions or variables when the package is imported.

> **Equivalent Import**

```python
import package.maths

package.maths.addition(2, 3)
```
