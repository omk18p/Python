## `array` Module

The `array` module provides **memory-efficient arrays** that store elements of the **same data type**.

### Example

```python
import array

arr = array.array('i', [1, 2, 3, 4])

print(arr)
```

- `import array` → Imports the built-in `array` module.
- `array.array()` → Creates an array.
- `'i'` → Type code for **integer**.
- `[1, 2, 3, 4]` → Initial elements.

Output

```python
array('i', [1, 2, 3, 4])
```

> **Note:** Unlike Python lists, arrays can store **only one data type**.

## Data Serialization (`json`)

**Serialization** converts a Python object into **JSON format**.  
**Deserialization** converts JSON back into a Python object.

### Example

```python
import json

data = {'name': 'Krish', 'age': 25}
```

### `json.dumps()`

Converts a **Python object** into a **JSON string**.

```python
json_str = json.dumps(data)
```

Output

```python
'{"name": "Krish", "age": 25}'
```

Type

```python
<class 'str'>
```

### `json.loads()`

Converts a **JSON string** back into a **Python object**.

```python
parsed_data = json.loads(json_str)
```

Output

```python
{'name': 'Krish', 'age': 25}
```

Type

```python
<class 'dict'>
```

> **Remember:**  
> `dumps()` → **Python → JSON String**  
> `loads()` → **JSON String → Python**

````python
## Data Serialization
import json
data={'name':'Krish','age':25}

json_str=json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
````

## Regular Expressions (`re`)

The `re` module is used to **search, match, and manipulate text using patterns**.

### Example

```python
import re

pattern = r'\d+'
text = 'There are 123 apples 456'

match = re.search(pattern, text)
print(match.group())
```

- `import re` → Imports the Regular Expression module.
- `r'...'` → Raw string (treats `\` as a literal backslash).
- `\d` → Matches a **digit** (`0–9`).
- `+` → Matches **one or more** occurrences.
- `re.search()` → Finds the **first match** in the string.
- `match.group()` → Returns the matched text.

Output

```python
123
```

> **Remember:**  
> `\d` → Digit (`0–9`)  
> `+` → One or more occurrences  
> `\d+` → One or more consecutive digits
