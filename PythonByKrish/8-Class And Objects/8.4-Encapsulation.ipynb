#### Encapsulation And Abstraction
Encapsulation and abstraction are two fundamental principles of Object-Oriented Programming (OOP) that help in designing robust, maintainable, and reusable code. Encapsulation involves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.

##### Encapsulation
Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data.


## Encapsulation (Public Variables)

**Encapsulation** is the process of **bundling data (variables) and methods** into a single class and controlling access to the data.

### Key Points

- `Person` → Class containing data and methods.
- `self.name` → **Public** instance variable (accessible from anywhere).
- `self.age` → **Public** instance variable.
- `get_name(person)` → Returns the public variable `name`.
- `person = Person("Krish", 34)` → Creates a `Person` object.
- `dir(person)` → Displays all attributes and methods of the object.

### Access Modifiers in Python

- **Public** → `name` (accessible everywhere)
- **Protected** → `_name` (convention: intended for internal use)
- **Private** → `__name` (name mangling restricts direct access)

> **Remember:** Python doesn't enforce strict access modifiers like Java or C++; it relies mainly on naming conventions.

````python
### Encapsulation  with Getter and Setter MEthods
### Public,protected,private variables or access modifiers

class Person:
    def __init__(self,name,age):
        self.name=name    ## public variables
        self.age=age      ## public variables

def get_name(person):
    return person.name

person=Person("Krish",34)
get_name(person)
````

````python
dir(person)
````

````python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name    ## private variables
        self.__age=age      ## private variables
        self.gender=gender

def get_name(person):
    return person.__name

person=Person("Krish",34,"Male")
get_name(person)##private variables cannot be accessed outside the class
````

## Encapsulation (Private Variables)

Private variables are used to **hide data** from direct access outside the class.

### Key Points

- `self.__name` → **Private** variable.
- `self.__age` → **Private** variable.
- `self.gender` → **Public** variable.
- Private variables can be accessed **only inside the class**.
- `get_name(person)` → Tries to access `__name` outside the class, which raises an error.
- To access private variables, use **getter** and **setter** methods defined inside the class.

### Access Modifiers

- **Public** → `name`
- **Protected** → `_name`
- **Private** → `__name`

> **Remember:** Private variables (`__variable`) cannot be accessed directly outside the class due to **name mangling**.

````python
dir(person)
````

````python
class Person:
    def __init__(self,name,age,gender):
        self._name=name    ## protected variables
        self._age=age      ## protected variables
        self.gender=gender

class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)


employee=Employee("Omkar",22,"Male")
print(employee._name)
````

## Encapsulation (Protected Variables)

Protected variables are intended to be accessed **within the class and its child classes**.

### Key Points

- `self._name` → **Protected** variable.
- `self._age` → **Protected** variable.
- `self.gender` → **Public** variable.
- `Employee` inherits from `Person` using `super()`.
- Protected variables **can be accessed** inside the child class.
- `employee._name` works, but accessing protected variables directly from outside the class is **discouraged** (by convention).

### Access Modifiers

- **Public** → `name`
- **Protected** → `_name`
- **Private** → `__name`

> **Remember:** Protected variables (`_variable`) are **accessible** outside the class but are intended for internal use by the class and its subclasses.

````python
## Encapsulation With Getter And Setter
class Person:
    def __init__(self,name,age):
        self.__name=name  ## Private access modifier or variable
        self.__age=age ## Private variable

    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## setter method for name
    def set_name(self,name):
        self.__name=name

    # Getter method for age
    def get_age(self):
        return self.__age
    
    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative.")


person=Person("Omkar",22)

## Access and modify private variables using getter and setter

print(person.get_name())
print(person.get_age())

person.set_age(35)
print(person.get_age())

person.set_age(-5)
    
````

## Encapsulation with Getter and Setter

**Getter** and **Setter** methods provide **controlled access** to private variables.

### Key Points

- `self.__name` and `self.__age` → **Private** variables.
- `get_name()` → Returns the value of `__name`.
- `set_name()` → Updates the value of `__name`.
- `get_age()` → Returns the value of `__age`.
- `set_age()` → Updates `__age` only if the value is valid (`age > 0`).
- Validation logic inside the setter prevents invalid data from being stored.
- Private variables are accessed **only through getter and setter methods**.

### Advantages

- Protects data from direct modification.
- Allows validation before updating values.
- Improves data security and maintainability.
