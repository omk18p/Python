##### Operator Overloading
Operator overloading allows you to define the behavior of operators (+, -, *, etc.) for custom objects. You achieve this by overriding specific magic methods in your class.

````python
#### Common Operator Overloading Magic Methods
'''
__add__(self, other): Adds two objects using the + operator.
__sub__(self, other): Subtracts two objects using the - operator.
__mul__(self, other): Multiplies two objects using the * operator.
__truediv__(self, other): Divides two objects using the / operator.
__eq__(self, other): Checks if two objects are equal using the == operator.
__lt__(self, other): Checks if one object is less than another using the < operator.

__gt__
'''
````

## Operator Overloading Magic Methods

These magic methods define how **operators behave for custom objects**.

### Common Magic Methods

- `__add__(self, other)` → `+` (Addition)
- `__sub__(self, other)` → `-` (Subtraction)
- `__mul__(self, other)` → `*` (Multiplication)
- `__truediv__(self, other)` → `/` (Division)
- `__eq__(self, other)` → `==` (Equality)
- `__lt__(self, other)` → `<` (Less than)
- `__gt__(self, other)` → `>` (Greater than)

### Advantages

- Enables **operator overloading** for custom classes.
- Makes custom objects behave like built-in data types.
- Improves code readability and flexibility.

> **Remember:** Python automatically calls these methods when the corresponding operator is used on an object.

````python
### MAthematical operation for vectors
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, other):
        return Vector(self.x * other, self.y * other)
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
## create objectss of the Vector Class
v1=Vector(2,3)
v2=Vector(4,5)

print(v1+v2)
print(v1-v2)
print(v1*3)
print(v1.__add__(v2))#internal call to the __add__ method
````

## Operator Overloading Example (Vector)

Operator overloading allows **custom objects** to use operators like `+`, `-`, `*`, and `==`.

### Code Explanation

- `Vector` → Class representing a 2D vector (`x`, `y`).
- `__init__()` → Initializes the vector coordinates.
- `__add__(other)` → Defines behavior of the `+` operator (adds two vectors).
- `__sub__(other)` → Defines behavior of the `-` operator (subtracts two vectors).
- `__mul__(other)` → Defines behavior of the `*` operator (scalar multiplication).
- `__eq__(other)` → Defines behavior of the `==` operator (compares two vectors).
- `__repr__()` → Returns the official string representation of the object.
- `v1 + v2` → Internally calls `v1.__add__(v2)`.
- `v1 - v2` → Internally calls `v1.__sub__(v2)`.
- `v1 * 3` → Internally calls `v1.__mul__(3)`.

### Output

```python
Vector(6, 8)
Vector(-2, -2)
Vector(6, 9)
```

> **Remember:** Operator overloading lets custom objects behave like built-in types by defining magic methods such as `__add__`, `__sub__`, `__mul__`, and `__eq__`.

````python
### Overloading Operators for Complex Numbers

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real_part, imag_part)

    def __truediv__(self, other):
        denominator = other.real**2 + other.imag**2
        real_part = (self.real * other.real + self.imag * other.imag) / denominator
        imag_part = (self.imag * other.real - self.real * other.imag) / denominator
        return ComplexNumber(real_part, imag_part)

    def __eq__(self, other):
        return self.real == other.real and self.imag == other.imag

    def __repr__(self):
        return f"{self.real} + {self.imag}i"

# Create objects of the ComplexNumber class
c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(1, 4)

# Use overloaded operators
print(c1 + c2)  # Output: 3 + 7i
print(c1 - c2)  # Output: 1 - 1i
print(c1 * c2)  # Output: -10 + 11i
print(c1 / c2)  # Output: 0.8235294117647058 - 0.29411764705882354i
print(c1 == c2) # Output: False
````

## Operator Overloading (Complex Numbers)

Operator overloading allows **complex number objects** to use arithmetic operators like built-in numeric types.

### Code Explanation

- `ComplexNumber` → Class representing a complex number (`real + imagi`).
- `__init__()` → Initializes the real and imaginary parts.
- `__add__(other)` → Adds two complex numbers using `+`.
- `__sub__(other)` → Subtracts two complex numbers using `-`.
- `__mul__(other)` → Multiplies two complex numbers using the complex multiplication formula.
- `__truediv__(other)` → Divides two complex numbers using the complex division formula.
- `__eq__(other)` → Compares two complex numbers using `==`.
- `__repr__()` → Returns a readable string representation.
- `c1 + c2` → Internally calls `c1.__add__(c2)`.
- `c1 - c2` → Internally calls `c1.__sub__(c2)`.
- `c1 * c2` → Internally calls `c1.__mul__(c2)`.
- `c1 / c2` → Internally calls `c1.__truediv__(c2)`.
- `c1 == c2` → Internally calls `c1.__eq__(c2)`.

### Output

```python
3 + 7i
1 + -1i
-10 + 11i
0.8235294117647058 + -0.29411764705882354i
False
```

> **Remember:** By implementing magic methods, custom classes can support arithmetic and comparison operators just like Python's built-in numeric types.
