#### Magic Methods
Magic methods in Python, also known as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behavior of objects for built-in operations, such as arithmetic operations, comparisons, and more.

## Magic (Dunder) Methods

Magic methods are **special methods** (starting and ending with `__`) that customize the behavior of objects.

### Common Magic Methods

- `__init__` → Initializes a new object (constructor).
- `__str__` → Returns a user-friendly string representation (`print(object)`).
- `__repr__` → Returns the official string representation (used for debugging).
- `__len__` → Returns the length of an object (`len(object)`).
- `__getitem__` → Retrieves an item using indexing (`obj[index]`).
- `__setitem__` → Sets or updates an item using indexing (`obj[index] = value`).

### Advantages

- Makes custom objects behave like built-in Python objects.
- Improves readability and usability.
- Enables operator overloading and custom behavior.

> **Remember:** Magic methods are automatically called by Python when performing operations like `print()`, `len()`, indexing, etc.

````python
## Basics MEthods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
    
person=Person("KRish",34)
print(person)
print(repr(person))
````

## `__str__` and `__repr__`

These magic methods define **how an object is represented as a string**.

### Key Points

- `__init__()` → Initializes the object's instance variables.
- `__str__()` → Returns a **user-friendly** string representation. Called by `print(object)`.
- `__repr__()` → Returns the **official/developer-friendly** string representation. Called by `repr(object)`.
- `print(person)` → Internally calls `person.__str__()`.
- `repr(person)` → Internally calls `person.__repr__()`.

### Output

```python
KRish,34 years old
Person(name=KRish,age=34)
```

> **Remember:**  
> `__str__` → For **users** (readable output).  
> `__repr__` → For **developers** (debugging/official representation).
