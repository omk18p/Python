## Instance Variables and Objects

An **instance variable** is a variable that belongs to a specific object.

### Code Explanation

- `class Dog:` → Defines a class named `Dog`.
- `__init__()` → Constructor; automatically called when an object is created.
- `self` → Refers to the current object (instance).
- `name, age` → Parameters passed while creating the object.
- `self.name = name` → Creates an instance variable `name`.
- `self.age = age` → Creates an instance variable `age`.
- `dog1 = Dog("Buddy", 3)` → Creates an object named `dog1`.
- `print(dog1)` → Prints the object's memory address (default representation).
- `print(dog1.name)` → Accesses and prints the object's `name`.
- `print(dog1.age)` → Accesses and prints the object's `age`.

### Output

```python
<__main__.Dog object at ...>
Buddy
3
```

> **Remember:**  
> - **Class** → Blueprint.  
> - **Object (Instance)** → Real object created from the class.  
> - **Instance Variables** → Variables unique to each object.

## `self` in Python

- `self` refers to the **current object (instance)**.
- It is used to access **instance variables** and **methods**.
- The first parameter of an instance method is **required**, but its name **does not have to be `self`**.
- By convention, Python programmers always use `self`.

> **Remember:** `self` is **not a keyword**; it is a naming convention.

````python
### Instance Variable and Methods
class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

## create objects
dog1=Dog("Buddy",3)

print(dog1)
print(dog1.name)
print(dog1.age)
    
    
````

````python
dog2=Dog("Lucy",4)
print(dog2.name)
print(dog2.age)
````

````python
## Define a class with instance methods
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} says woof")


dog1=Dog("Buddy",3)
dog1.bark()
dog2=Dog("Lucy",4)
dog2.bark()

````

## Instance Methods

An **instance method** operates on a specific object and can access its instance variables using `self`.

### Code Explanation

- `bark()` → Instance method of the `Dog` class.
- `self.name` → Accesses the `name` of the current object.
- `dog1.bark()` → Calls `bark()` for `dog1`.
- `dog2.bark()` → Calls `bark()` for `dog2`.
- Each object uses **its own** `name` value.

### Output

```python
Buddy says woof
Lucy says woof
```

> **Remember:** Instance methods use `self` to access the current object's data.

````python
### Modeling a Bank Account

## Define a class for bank account
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient funds!")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New Balance is {self.balance}")

    def get_balance(self):
        return self.balance
    
## create an account

account=BankAccount("Krish",5000)
print(account.balance)

    
````

````python
## Call isntance methods
account.deposit(100)
````

````python
account.withdraw(300)
````

````python
print(account.get_balance())
````
