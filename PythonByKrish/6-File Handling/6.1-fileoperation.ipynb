#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

## Reading a Whole File

Reads the **entire contents** of a file at once.

### Example

```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

- `open('example.txt', 'r')` → Opens the file in **read mode**.
- `with` → Automatically closes the file after use.
- `file.read()` → Reads the **entire file** as a single string.
- `content` → Stores the file contents.
- `print(content)` → Displays the file contents.

````python
### Read a Whole File

with open('example.txt','r') as file:
    content=file.read()
    print(content)
````

## Reading a File Line by Line

Reads a file **one line at a time**, which is memory-efficient for large files.

### Example

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

- `with` → Automatically closes the file.
- `for line in file` → Reads one line at a time.
- `strip()` → Removes leading/trailing whitespace, including the newline (`\n`) character.
- `print()` → Displays each cleaned line.

> **Note:** Reading line by line is preferred for **large files**.

````python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## sstrip() removes the newline character
````

## Writing to a File (Overwrite Mode)

Writes data to a file. If the file already exists, its contents are **overwritten**.

### Example

```python
with open('example.txt', 'w') as file:
    file.write('Hello World!\n')
    file.write('This is a new line.')
```

- `open('example.txt', 'w')` → Opens the file in **write mode**.
- `'w'` → Creates a new file if it doesn't exist; otherwise **overwrites** the existing file.
- `file.write()` → Writes text to the file.
- `\n` → Inserts a new line.
- `with` → Automatically closes the file.

> **Output in `example.txt`:**

```text
Hello World!
This is a new line.
```

````python
## Writing a file(Overwriting)

with open('example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('this is a new line.')
````

## Appending to a File (Without Overwriting)

Adds new data to the **end of a file** without removing existing contents.

### Example

```python
with open('example.txt', 'a') as file:
    file.write("Append operation taking place!\n")
```

- `open('example.txt', 'a')` → Opens the file in **append mode**.
- `'a'` → Adds new content to the end of the file.
- `file.write()` → Writes text to the file.
- `\n` → Inserts a new line.
- `with` → Automatically closes the file.

> **Note:** Unlike `'w'`, append mode (`'a'`) **does not overwrite** existing file contents.

````python
## Write a file(without Overwriting)
with open('example.txt','a') as file:
    file.write("\nAppend operation taking place!\n")
````

````python
### Writing a list of lines to a file
lines=['First line \n','Second line \n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
````

````python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
````

````python
# Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
````

````python
### Read the content froma  source text fiile and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
````

````python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
````

The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten).

````python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
````
