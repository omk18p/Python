#### The map() Function in Python
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

````python
def square(x):
    return x*x

square(10)
````

````python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
````

## `map()` with a Function

Applies a function to **every element** of an iterable.

```python
list(map(square, numbers))
```

Equivalent to:

```python
result = []
for num in numbers:
    result.append(square(num))
```

- `square` → Function applied to each element.
- `numbers` → Iterable.
- `map()` → Applies the function to every element.
- `list()` → Converts the map object into a list.

Output

```python
[1, 4, 9, 16, 25, 36]
```

````python
## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
````

````python
### MAp multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
````

## `map()` with Multiple Iterables

`map()` can apply a function to **multiple iterables** simultaneously.

### Syntax

```python
map(function, iterable1, iterable2, ...)
```

### Example

```python
added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
```

- `x` → Element from `numbers1`.
- `y` → Element from `numbers2`.
- `lambda x, y: x + y` → Adds corresponding elements.
- `map()` → Applies the lambda to both lists.
- `list()` → Converts the map object into a list.

Output

```python
[5, 7, 9]
```

> **Equivalent to:**

```python
result = []
for x, y in zip(numbers1, numbers2):
    result.append(x + y)
```

````python
## map() to convert a list of strings to integers
# Use map to convert strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))

print(int_numbers)  # Output: [1, 2, 3, 4, 5]
````

````python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
````

````python
def get_name(person):
    return person['name']

people=[
    {'name':'Krish','age':32},
    {'name':'Jack','age':33}
]
list(map(get_name,people))

````
