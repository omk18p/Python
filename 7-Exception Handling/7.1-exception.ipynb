#### Understanding Exceptions

Exception handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exceptions, including how to use try, except, else, and finally blocks.

##### What Are Exceptions?
Exceptions are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:

- ZeroDivisionError: Dividing by zero.
- FileNotFoundError: File not found.
- ValueError: Invalid value.
- TypeError: Invalid type.

````python
## Exception try ,except block

try:
    a=b
except:
    print("The variable has not been assigned")
````

## `try` and `except`

Used to **handle errors (exceptions)** without stopping the program.

### Syntax

```python
try:
    # Code that may cause an error
except:
    # Code to execute if an error occurs
```

### Example

```python
try:
    a = b
except:
    print("The variable has not been assigned")
```

- `try` → Executes code that may raise an exception.
- `a = b` → Raises a `NameError` because `b` is not defined.
- `except` → Catches the exception and executes the block instead of terminating the program.

Output

```python
The variable has not been assigned
```

> **Remember:**  
> `try` → Test the code.  
> `except` → Handle the error if one occurs.

## Multiple `except` Blocks

Use multiple `except` blocks to handle **different types of exceptions** separately.

### Example

```python
try:
    result = 1 / 2
    a = b
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter the denominator greater than 0")
except Exception as ex1:
    print(ex1)
    print("Main exception got caught here")
```

- `try` → Executes code that may raise an exception.
- `except ZeroDivisionError as ex` → Handles only division-by-zero errors.
- `except Exception as ex1` → Handles any other exception.
- `as ex` / `as ex1` → Stores the exception object to access its error message.

### Output

```python
name 'b' is not defined
Main exception got caught here
```

> **Remember:**  
> - Handle **specific exceptions first**.  
> - `Exception` is the **generic (catch-all)** exception and should usually be placed last.

````python
try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
except Exception as ex:
    print(ex)
````

## `try`, `except`, `else`

The `else` block executes **only if no exception occurs** in the `try` block.

### Syntax

```python
try:
    # Risky code
except ExceptionType:
    # Handle exception
else:
    # Runs only if no exception occurs
```

### Example

```python
try:
    num = int(input("Enter a number:"))
    result = 10 / num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
```

- `try` → Executes code that may raise an exception.
- `except ValueError` → Handles invalid input.
- `except ZeroDivisionError` → Handles division by zero.
- `except Exception as ex` → Handles any other exception.
- `else` → Executes **only if no exception occurs**.

> **Remember:**  
> `try` → Test code.  
> `except` → Handle errors.  
> `else` → Runs only when `try` succeeds without errors.

````python
## try,except,else block
try:
    num=int(input("Enter a number:"))
    result=10/num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")

    

````

````python
## try,except,else and finally
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
finally:
    print("Execution complete.")

````

## `try`, `except`, `else`, `finally`

Used for **exception handling**.

### Code Explanation

- `num = int(input(...))` → Takes user input and converts it to an integer.
- `result = 10 / num` → Divides `10` by the entered number.
- `except ValueError` → Executes if the input is **not a valid integer**.
- `except ZeroDivisionError` → Executes if the user enters `0`.
- `except Exception as ex` → Handles any other unexpected exception.
- `else` → Executes only if the `try` block completes without any exception and prints the result.
- `finally` → Always executes, regardless of whether an exception occurs or not.

> **Remember:**  
> `try` → Test code.  
> `except` → Handle errors.  
> `else` → Runs if no error occurs.  
> `finally` → Always runs (commonly used for cleanup, e.g., closing files).

````python
### File handling and Exception HAndling

try:
    file=open('example1.txt','r')
    content=file.read()
    a=b
    print(content)

except FileNotFoundError:
    print("The file does not exists")
except Exception as ex:
    print(ex)

finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print('file close')
````
